We can use np.where() to get the indices where the elements with a Numpy array meet a condition

```python
print(my_array)
indices = np.where(my_array % 2 == 0)  #it will return the index values
print(indices)
print(my_array)
print(my_array[indices])

#or
print(my_array[np.where(my_array % 2 == 0)])
```

```python
#We can also use np.where() to modify Numpy array
print(np.where(my_array > 2, 100, my_array))  #(Condition, Case_if_met, else); does not act in place
#note: The original array remains same
print(my_array)
```

```python
new_array = np.where(my_array < 4, my_array ** 2, -55)
print(new_array)
print(my_array)
```

We can also directly modify Numpy arrays using subscripts

```python
my_array = np.array([1,2,3,4,5])
my_array_2  = my_array              #Making Shallow copy
my_array_2
```

```python
my_array_2[my_array_2 > 2]=100      #If anything is changed in my_array_2 that changes apply to my_array
print(my_array_2)
```

```python
my_array
```

```python
my_array = np.array([1,2,3,4,5])
my_array_2 = my_array.copy()                #Making Deep copy
my_array_2
```

```python
my_array_2[my_array_2 > 2 ] = 100
my_array_2
```

```python
my_array                                    #my_array_2 is a deep copy, therefore my_array is not modified
```
